```python
import plotly.offline as offline
offline.init_notebook_mode(connected=True)
```

```python
import plotly.graph_objs as go

import pandas as pd
import numpy as np
```

```python
n= 200

random_x =np.random.randn(n)

random_y =np.random.randn(n)

```

```python
trace =go.Scatter(
    x= random_x,
    y= random_y,
    mode = 'markers'
)

trace
```

```python
data = [trace]
offline.iplot(data)
```

```python
n=50
random_x=np.linspace(0,1,n)
```

```python
random_y0=np.random.randn(n)+5
random_y1=np.random.randn(n)
random_y2=np.random.randn(n)-5
```

```python
trace0=go.Scatter(
    x= random_x,
    y=random_y0,
    mode='markers'

)
```

```python
trace1=go.Scatter(
    x= random_x,
    y=random_y1,
    mode='lines'

)
```

```python
trace2=go.Scatter(
    x= random_x,
    y=random_y2,
    mode='lines+markers'

)
```

```python
data = [trace0,trace1,trace2]
offline.iplot(data)
```

```python
honey_data = pd.read_csv(r'C:\Users\JawadSajid\Desktop\DV\All Online Lectures\Lectures Week 2\Lectures\honeyproduction.csv')
```

```python
honey_data.head()
```

```python
trace=go.Scatter(
    x= honey_data['state'],
    y= honey_data['year'],
    mode='markers',
    marker=dict(
        size=14,
        color=honey_data['totalprod'],
        colorscale='Rainbow',
        showscale=True,
        opacity=0.5
    ),
    text=honey_data['totalprod']
)
```

```python
data=[trace]
```

```python
layout=go.Layout(
    title= 'Honey production United States',
    hovermode='closest',
    xaxis=dict(title='States'),
    yaxis=dict(title='Boney Production')
)
```

```python
fig = dict(data=data,layout=layout)
```

```python
offline.iplot(fig)
```
